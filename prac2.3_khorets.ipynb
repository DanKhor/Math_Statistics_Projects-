```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt 
```

```python
class radius_gen(sts.rv_continuous):
    def _pdf(self, x):
        return 2*x
radius = radius_gen(a=0, b=1, name='radius')
```

```python
n = 10000
r = radius.rvs(size=n)
phi = sts.uniform.rvs(loc=0, scale=2*np.pi, size=n)
X = r * np.cos(phi)
Y = r * np.sin(phi)
```

```python
circle = plt.Circle((0, 0), 1, color='g', alpha=0.5)
fig, ax = plt.subplots()
ax.add_artist(circle)
plt.scatter(X, Y, color='b', s=0.4)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.show()
```
