```python
import scipy.stats as sts
import numpy as np 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')

TET = 2 
TET1 = (5, 1)
n = 100
domain = np.linspace(-3, 3, 100)
```

Критическое множество вида sum(Xi)/n > C

```python
def error1(C, T):
    return 1 - sts.gamma.cdf(n*C, n, scale=1/T)
def error2(C, T):
    return sts.gamma.cdf(n*C, n, scale=1/T)
```

```python
for T1 in TET1:
    fig, ax = plt.subplots()
    ax.plot(domain, error1(domain, TET), label='error 1')
    ax.plot(domain, error2(domain, T1), label='error 2')
    ax.legend()
    ax.set_title('TETA = 2, T1 = ' + str(T1))
    ax.set_xlabel('C')
    ax.set_ylabel('probability')
    plt.show()
```

Критическое множество вида sum(Xi)/n < C

```python
for T1 in TET1:
    fig, ax = plt.subplots()
    ax.plot(domain, error2(domain, TET), label='error 1')
    ax.plot(domain, error1(domain, T1), label='error 2')
    ax.legend()
    ax.set_title('TETA = 2, T1 = ' + str(T1))
    ax.set_xlabel('C')
    ax.set_ylabel('probability')
    plt.show()
```

Критическое множество вида min(Xi) < C

```python
def error1_m(C, T):
    return sts.expon.cdf(n*C, scale=1/T)
def error2_m(C, T):
    return 1 - sts.expon.cdf(n*C, scale=1/T)
```

```python
for T1 in TET1:
    fig, ax = plt.subplots()
    ax.plot(domain, error1_m(domain, TET), label='error 1')
    ax.plot(domain, error2_m(domain, T1), label='error 2')
    ax.legend()
    ax.set_title('TETA = 2, T1 = ' + str(T1))
    ax.set_xlabel('C')
    ax.set_ylabel('probability')
    plt.show()
```
