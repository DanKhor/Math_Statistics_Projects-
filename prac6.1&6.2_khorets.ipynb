```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.stats as sts
import scipy.special as sps

n_tup = (5, 10, 20, 50, 100)
domain = np.linspace(-2, 2, 1000)
TETA = 0.4
```

```python
def apost_density(u, n, sample):
    s = sum(sample)
    return (u**(s) * (1-u)**(n-s) * sts.uniform.pdf(u, loc=0, scale=1))/sps.beta(1+s, n+1-s)

def apost_sum(u, n, s):
    return (u**(s) * (1-u)**(n-s) * sts.uniform.pdf(u, loc=0, scale=1))/sps.beta(1+s, n+1-s)
```

```python
%%capture --no-display
#6.1 X1, ..., Xn ∼ Bern(θ), θ ∼ R[0, 1]. Построить график апостериорной плотности θ
for n in n_tup:
    sample = sts.bernoulli.rvs(TETA, size=n)
    figure = plt.figure(figsize=(16,10))
    figure.tight_layout(pad=7.0)
    plt.subplot(1, 3, 1)
    plt.plot(domain, apost_density(domain, n, sample))
    plt.title('Aposterior density of U[0,1] given Bernoullis, n = ' + str(n))
    plt.xlabel('Value')
    plt.ylabel('Density')
    
    plt.subplot(1, 3, 2)
    plt.plot(domain, apost_sum(domain, n, 9*n/10))
    plt.title('Aposterior density of U[0,1] sum = 9*n/10, n = ' + str(n))
    plt.xlabel('Value')
    plt.ylabel('Density')
    
    plt.subplot(1, 3, 3)
    plt.plot(domain, apost_sum(domain, n, 99*n/100))
    plt.title('Aposterior density of U[0,1] sum = 99*n/100, n = ' + str(n))
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
```

```python
a = 2
b = 3
def apost_density_bet(u, n, sample):
    s = sum(sample)
    return (u**(s) * (1-u)**(n-s) * sts.beta.pdf(u, a, b, loc=0, scale=1))/sps.beta(1+s, n+1-s)

def apost_sum_bet(u, n, s):
    return (u**(s) * (1-u)**(n-s) * sts.beta.pdf(u, a, b, loc=0, scale=1))/sps.beta(1+s, n+1-s)
```

```python
%%capture --no-display
#Сделать то же, что в 6.1 для X1, ..., Xn ∼ Bern(θ), θ ∼ Beta(a, b).
for n in n_tup:
    sample = sts.bernoulli.rvs(TETA, size=n)
    figure = plt.figure(figsize=(16,10))
    figure.tight_layout(pad=7.0)
    plt.subplot(1, 3, 1)
    plt.plot(domain, apost_density_bet(domain, n, sample))
    plt.title('Aposterior density of B(a,b) given Bernoullis, n = ' + str(n))
    plt.xlabel('Value')
    plt.ylabel('Density')
    
    plt.subplot(1, 3, 2)
    plt.plot(domain, apost_sum_bet(domain, n, 9*n/10))
    plt.title('Aposterior density of B(a,b) sum = 9*n/10, n = ' + str(n))
    plt.xlabel('Value')
    plt.ylabel('Density')
    
    plt.subplot(1, 3, 3)
    plt.plot(domain, apost_sum_bet(domain, n, 99*n/100))
    plt.title('Aposterior density of B(a,b) sum = 99*n/100, n = ' + str(n))
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
```
