```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stat

binom = stat.binom.rvs(100, 0.04, size=1000)
poiss = stat.poisson.rvs(4, size=1000)

#1) Моделировать 1000 реализаций с.в. X ∼ Bin(100, 0.04) и с.в. Y ∼ P oisss(4). Построить
#столбцовые диаграммы отдельно и на одном графике, сравнить. То же для X ∼ Bin(1000, 0.004) и
#X ∼ Bin(10, 0.4).


plt.hist(binom, bins=25)
plt.show()
plt.hist(poiss, bins=25)
plt.show()
#Теперь вместе
plt.hist([binom, poiss], bins=25, label=['bin(100, 0.04)', 'poiss'])
plt.legend()
plt.show()
```

```python
binom = stat.binom.rvs(1000, 0.004, size=1000)
plt.hist(binom, bins=25)
plt.show()
plt.hist(poiss, bins=25)
plt.show()
#Теперь вместе
plt.hist([binom, poiss], bins=25, label=['bin(1000, 0.004)', 'poiss'])
plt.legend()
plt.show()
```

```python
binom = stat.binom.rvs(10, 0.4, size=1000)
plt.hist(binom, bins=25)
plt.show()
plt.hist(poiss, bins=25)
plt.show()
#Теперь вместе
plt.hist([binom, poiss], bins=25, label=['bin(10, 0.4)', 'poiss'])
plt.legend()
plt.show()
```
