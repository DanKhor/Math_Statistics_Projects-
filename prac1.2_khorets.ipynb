```python
import matplotlib.pyplot as plt 
import numpy as np 
import scipy.stats as stats
import scipy.special as sp

domain = np.linspace(-20, 20, 1000)


def sum_pdf(n, x):
    helpme = [sp.binom(n, k) * stats.gamma.pdf(x - n + k, n, loc=0, scale=1) for k in range(n+1)]
    return ((0.5)**n)*sum(helpme)

for n in range(1, 11):
    plt.plot(domain, sum_pdf(n, domain), label=str(n))
plt.title('n from 1 to 10')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()    
```

```python
#Expection equals 1.5*n , dispersion - 1.25*n 
def sum_center_norm_pdf(n, x):
    return ((1.25*n)**(0.5)) * sum_pdf(n, x * ((1.25*n)**(0.5)) + 1.5 * n)

for n in range(1, 11):
    plt.plot(domain, sum_center_norm_pdf(n, domain), label='sum_center_norm')
    plt.plot(domain, stats.norm.pdf(domain, 0, 1), label='Normal(0, 1)')
    plt.title('n = ' + str(n))
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```
