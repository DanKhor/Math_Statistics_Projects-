```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.stats as sts
from sklearn.linear_model import LinearRegression
plt.style.use('ggplot')
```

```python
a = 3
b = 2
m_tup = (10, 20, 50)
sigm_tup = (1, a/9)
```

```python
def MSE(y, y_pred):
    return np.mean((y - y_pred)**2)
```

<h1> Задача №11.1

```python
for m in m_tup:
    for sigm in sigm_tup:
        pts = LinearRegression()
        fig, ax = plt.subplots()
        
        X = np.array([k/m for k in range(0, m)])
        eps = sts.norm.rvs(loc=0, scale=sigm, size=m)
        y = a*X + b + eps
        
        #Облако точек
        ax.scatter(X, y, c='g')
        
        #Исходная прямая 
        ax.plot(X, a*X + b, label='initial')
        
        #Обучим модель 
        X = np.array([X]).T
        pts.fit(X, y) 
        
        y_pred = pts.predict(X)
        
        #Mean Square Error
        print('MSE:', MSE(y, y_pred))
        
        #R^2 
        print('R^2:', pts.score(X, y))
        
        ax.plot(X, y_pred, label='predict')
        
        print('b_est =', pts.intercept_)
        print('a_est =', pts.coef_[0])
        
        ax.legend()
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_title('Regression, m = ' + str(m) + ', sigma = ' + str(round(sigm, 2))) 
        plt.show()
```

<h1> Задача №11.2 </h1>
 1) Отклоняется одна точка 

```python
for m in m_tup:
    t = int(m/2) 
    for sigm in sigm_tup:
        pts = LinearRegression()
        fig, ax = plt.subplots()
        
        X = np.array([k/m for k in range(0, m)])
        eps = sts.norm.rvs(loc=0, scale=sigm, size=m)
        y = a*X + b + eps
        y[t] = a*X[t] + b + 5*sigm
        
        #Облако точек
        ax.scatter(X, y, c='g')
        
        #Исходная прямая 
        ax.plot(X, a*X + b, label='initial')
        
        #Обучим модель 
        X = np.array([X]).T
        pts.fit(X, y) 
        
        y_pred = pts.predict(X)
        
        #Mean Square Error
        print('MSE:', MSE(y, y_pred))
        
        #R^2 
        print('R^2:', pts.score(X, y))
        
        ax.plot(X, y_pred, label='predict')
        
        print('b_est =', pts.intercept_)
        print('a_est =', pts.coef_[0])
        
        ax.legend()
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_title('Regression, m = ' + str(m) + ', sigma = ' + str(round(sigm, 2))) 
        plt.show()
```

2) Отклоняются две точки

```python
for m in m_tup:
    t = int(m/3) 
    s = int(2*m/3)
    for sigm in sigm_tup:
        pts = LinearRegression()
        fig, ax = plt.subplots()
        
        X = np.array([k/m for k in range(0, m)])
        eps = sts.norm.rvs(loc=0, scale=sigm, size=m)
        y = a*X + b + eps
        y[t] = a*X[t] + b + 5*sigm
        y[s] = a*X[s] + b - 5*sigm
        
        #Облако точек
        ax.scatter(X, y, c='g')
        
        #Исходная прямая 
        ax.plot(X, a*X + b, label='initial')
        
        #Обучим модель 
        X = np.array([X]).T
        pts.fit(X, y) 
        
        y_pred = pts.predict(X)
        
        #Mean Square Error
        print('MSE:', MSE(y, y_pred))
        
        #R^2 
        print('R^2:', pts.score(X, y))
        
        ax.plot(X, y_pred, label='predict')
        
        print('b_est =', pts.intercept_)
        print('a_est =', pts.coef_[0])
        
        ax.legend()
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_title('Regression, m = ' + str(m) + ', sigma = ' + str(round(sigm, 2))) 
        plt.show()
```
