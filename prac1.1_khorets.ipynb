```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import scipy.special as sp
import math

def cdf_gamma_center_norm(x, n):
    return (sp.gammainc(n, x * ((n)**(1/2)) + n))

def pdf_gamma_center_norm(x, n):
    return (n**(1/2) * (x * n**(1/2) + n)**(n-1) * np.exp(-n**(1/2) * x - n))/sp.gamma(n)
domain = np.linspace(0,5, 1000)

#Построим графики плотностей Sn:
for n in range(1,11):
    plt.plot(domain, stats.gamma.pdf(domain, n, loc=0, scale=1))
    plt.title('Sum of '+str(n) + ' values')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
```

```python
#Изобразить на одном графике плотность центрированной и нормированной с.в. Sn и плотность N(0, 1)
#для n = 1, 2, . . . , 20. Также изобразить графики ф.р.
for n in range(1,21):
    plt.plot(domain, stats.norm.pdf(domain, 0, 1), 'r', label='norm')
    plt.plot(domain, pdf_gamma_center_norm(domain, n), label='gamma_center_norm')
    plt.title('n = ' + str(n))
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```

```python
#Теперь функции распределний
domain = np.linspace(-20,20, 1000)
for n in range(1,21):
    plt.plot(domain, stats.norm.cdf(domain, 0, 1), 'r', label='norm')
    plt.plot(domain, cdf_gamma_center_norm(domain, n), label='gamma_center_norm')
    plt.xlabel('Value')
    plt.title('n = '+ str(n))
    plt.ylabel('Prob')
    plt.legend()
    plt.show()
```
