```python
import numpy as np 
import scipy.stats as sts 
import matplotlib.pyplot as plt 

N = 10000
```

```python
x = sts.uniform.rvs(loc=-((2)**(0.5))/2, scale=(2)**(0.5), size=N)
y = sts.uniform.rvs(loc=-((2)**(0.5))/2, scale=(2)**(0.5), size=N)
```

```python
x1 = (((2)**(0.5))/2)*x + (((2)**(0.5))/2)*y
y1 = -(((2)**(0.5))/2)*x + (((2)**(0.5))/2)*y
```

```python
domain = np.linspace(-1, 1, 2)
```

```python
plt.scatter(x1, y1, s=0.5)
plt.ylim(0,1)
plt.xlim(-1,1)
plt.plot(domain, domain + 1, '-r')
plt.plot(domain, 1 - domain, '-r')
plt.hlines(0, -1, 1, 'r')
plt.show()
```
