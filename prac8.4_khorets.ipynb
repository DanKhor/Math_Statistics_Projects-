```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt 
plt.style.use('ggplot')

n_tup = (1, 5, 20, 50, 100, 500)
T0 = 1/2
T1 = 2/3
C = np.linspace(-3, 5, 500)
```

Sum(Xi) ~ NB(n, T)
При ppf(1-a, T1)/C < n < ppf(a, T0)/C вероятности ошибок < a=0.05
Построим графики ошибок первого и второго рода:

```python
for n in n_tup:
    fig, ax = plt.subplots(figsize=(7, 5))
    ax.plot(C, sts.nbinom.cdf(n*C, n, T0, loc=0), label='1 рода')
    ax.plot(C, 1 - sts.nbinom.cdf(n*C, n, T1, loc=0), label='2 рода')
    ax.hlines(0.05, -3, 5, label='alpha = 0.05')
    ax.set_xlabel('C')
    ax.set_title('n = ' + str(n))
    ax.legend()
    plt.show()
```
