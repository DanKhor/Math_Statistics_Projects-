```python
import numpy as np
import scipy.stats as sts 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')

T0 = 2
T1 = (5, 1)
n = 100
m = 1000
OY_pt = [i/m for i in range(1, m+1)]
domain = np.linspace(0, 1, 2)
```

Для критического множества sum(Xi)/n > C 

```python
pvalue_0_1 = []
pvalue_1_1 = []
pvalue_2_1 = []
for i in range(m):
    
    sample= sts.expon.rvs(scale=1/T0, size=n)
    T = np.mean(sample)
    # p_value hypotese: 
    p0 = 1 - sts.gamma.cdf(n*T, n, scale=1/T0)
    pvalue_0_1.append(p0)
    
    # p_value alternative1:
    sample = sts.expon.rvs(scale=1/T1[0], size=n)
    T = np.mean(sample)
    p1 = 1- sts.gamma.cdf(n*T, n, scale=1/T0)
    pvalue_1_1.append(p1)
    
    # p_value alternative2:
    sample = sts.expon.rvs(scale=1/T1[1], size=n)
    T = np.mean(sample)
    p2 = 1 - sts.gamma.cdf(n*T, n, scale=1/T0)
    pvalue_2_1.append(p2)

pvalue_0_1.sort()
pvalue_1_1.sort()
pvalue_2_1.sort()
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(pvalue_0_1, OY_pt, label='p_value H0')
ax.plot(pvalue_1_1, OY_pt, label='p_value H1')
ax.plot(pvalue_2_1, OY_pt,color='g', label='p_value H2')
ax.plot(domain, domain, label='y = x')
ax.legend()
ax.set_title('Criteria: sum(Xi)/n > C')
plt.show()
```

Для критического множества sum(Xi)/n < C

```python
pvalue_0_2 = []
pvalue_1_2 = []
pvalue_2_2 = []
for i in range(m):
    
    sample= sts.expon.rvs(scale=1/T0, size=n)
    T = np.mean(sample)
    # p_value hypotese: 
    p0 = sts.gamma.cdf(n*T, n, scale=1/T0)
    pvalue_0_2.append(p0)
    
    # p_value alternative1:
    sample = sts.expon.rvs(scale=1/T1[0], size=n)
    T = np.mean(sample)
    p1 = sts.gamma.cdf(n*T, n, scale=1/T0)
    pvalue_1_2.append(p1)
    
    # p_value alternative2:
    sample = sts.expon.rvs(scale=1/T1[1], size=n)
    T = np.mean(sample)
    p2 = sts.gamma.cdf(n*T, n, scale=1/T0)
    pvalue_2_2.append(p2)

pvalue_0_2.sort()
pvalue_1_2.sort()
pvalue_2_2.sort()
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(pvalue_0_2, OY_pt, label='p_value H0')
ax.plot(pvalue_1_2, OY_pt, label='p_value H1')
ax.plot(pvalue_2_2, OY_pt,color='g', label='p_value H2')
ax.plot(domain, domain, label='y = x')
ax.legend()
ax.set_title('Criteria: sum(Xi)/n < C')
plt.show()
```

Для критического множества X(1) < C

```python
pvalue_0_3 = []
pvalue_1_3 = []
pvalue_2_3 = []
for i in range(m):
    
    sample= sts.expon.rvs(scale=1/T0, size=n)
    T = min(sample)
    # p_value hypotese: 
    p0 = sts.expon.cdf(n*T, scale=1/T0)
    pvalue_0_3.append(p0)
    
    # p_value alternative1:
    sample = sts.expon.rvs(scale=1/T1[0], size=n)
    T = min(sample)
    p1 = sts.expon.cdf(n*T, scale=1/T0)
    pvalue_1_3.append(p1)
    
    # p_value alternative2:
    sample = sts.expon.rvs(scale=1/T1[1], size=n)
    T = min(sample)
    p2 = sts.expon.cdf(n*T, scale=1/T0)
    pvalue_2_3.append(p2)

pvalue_0_3.sort()
pvalue_1_3.sort()
pvalue_2_3.sort()
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(pvalue_0_3, OY_pt, label='p_value H0')
ax.plot(pvalue_1_3, OY_pt, label='p_value H1')
ax.plot(pvalue_2_3, OY_pt,color='g', label='p_value H2')
ax.plot(domain, domain, label='y = x')
ax.legend()
ax.set_title('Criteria: X(1) < C')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(pvalue_0_1, OY_pt, label='p_value H0, 1crit')
ax.plot(pvalue_0_2, OY_pt, label='p_value H0, 2crit')
ax.plot(pvalue_0_3, OY_pt, label='p_value H0, 3crit')
ax.plot(pvalue_1_1, OY_pt, label='p_value H1, 1crit')
ax.plot(pvalue_1_2, OY_pt, label='p_value H1, 2crit')
ax.plot(pvalue_1_3, OY_pt, label='p_value H1, 3crit')
ax.plot(domain, domain, label='y = x')
ax.legend()
ax.set_title('Different criterions, hypotese H0, alternative: H1')
plt.show()
```
