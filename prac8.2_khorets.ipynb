```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts 
plt.style.use('ggplot')

a = 0.05
TET = 2
n_tup = (1, 5, 20, 50, 100, 500)
T1 = np.linspace(0.0001, 5, 1000)
```

```python
for n in n_tup:
    fig, ax = plt.subplots()
    #найдем C для первого критерия 
    C1 = sts.gamma.ppf(1-a, n, scale=1/TET)/n
    print('C1 =', C1)
    ax.plot(T1, 1 - sts.gamma.cdf(n*C1, n, scale=1/T1), label='beta(T1) for critic1')
    
    #найдем C для второго критерия
    C2 = sts.gamma.ppf(a, n, scale=1/TET)/n
    print('C2 =', C2)
    ax.plot(T1, sts.gamma.cdf(n*C2, n, scale=1/T1), label='beta(T1) for critic2')
    
    #найдем C для третьего критерия 
    C3 = sts.expon.ppf(a, scale=1/(n*TET))
    print('C3 =', C3)
    ax.plot(T1, sts.expon.cdf(n*C3, scale=1/(T1)), label='beta(T1) for critic3')
    
    ax.set_title('n = ' + str(n))
    ax.set_xlabel('T1')
    ax.set_ylabel('beta(T1)')
    ax.legend()
    plt.show()
```
