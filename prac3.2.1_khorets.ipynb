```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

tips = sns.load_dataset('tips')
tips['sex_smoker'] = tips["sex"].astype('str') + tips['smoker'].astype('str')
tips['sex_smoker'] = tips['sex_smoker'].astype('category')
tips.head()
```

```python
#1) построить диаграмму рассеяния для total_bill и tip, цветом задать пол, формой точек – время;
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='time')
plt.show()
```

```python
#2) сравнить курящих женщин, некурящих женщин, курящих мужчин и некурящих мужчин с помощью
#параллельных координат.

#tips_new = tips.loc[:,['total_bill', 'day', 'sex_smoker']]
#tips_new.head()
sns.lineplot(x='day', y='total_bill', hue='sex_smoker', data=tips)
sns.catplot(x='day', y='total_bill', hue='sex_smoker',kind='point', data=tips);
```

```python
#3) построить boxplot() для размера чаевых по дням недели.
box = sns.boxplot(x='day', y='tip', data=tips)
```
