```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts
from scipy.optimize import minimize

TETA = 1
n_tup = (5, 10, 20, 50, 100)
def MLE(x, sample):
            mle = 1
            for i in range(sample[len(sample)-1]):
                mle = mle*(1 + (sample[i]-x)**2)
            return mle
```

```python
#5.1
x0 = 0.0
df = pd.DataFrame(columns=['n', 'TETA_MLE_MEAN', 'TETA_MED_MEAN', 'TETA_MLE_VAR', 'TETA_MED_VAR'])
for n in n_tup:
    #def MLE(x, sample):
     #       mle = 1
      #      for i in range(n):
       #         mle = mle*(1 + (sample[i]-x)**2)
        #    return mle
    teta_mle = []
    teta_med = []
    for k in range(100):
        sample = sts.cauchy.rvs(loc=TETA, size=n)
        sample = sample.tolist()
        sample.append(n)
        res = minimize(MLE,x0, sample) #Ищем минимум знаменателя, чтобы получить максимум функции
        teta_mle.append(res.x[0])
        teta_med.append(np.median(sample))
#    np.array([n, np.mean(teta_mme), np.mean(teta_med), np.var(teta_mme), np.var(teta_med)])
    df_buf = pd.DataFrame({'n': [n], 'TETA_MLE_MEAN': [np.mean(teta_mle)], 
                           'TETA_MED_MEAN': [np.mean(teta_med)], 'TETA_MLE_VAR': [np.var(teta_mle)], 'TETA_MED_VAR':np.var(teta_med)})
    df = df.append(df_buf)
display(df)
```
